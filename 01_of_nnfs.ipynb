#  Implementing a Single Neuron from Scratch

## What is a Neuron?

- A **neuron** takes several inputs, multiplies them with their respective **weights**, adds a **bias**, and outputs a single value.
- This is the fundamental building block of a **neural network**.

---

## Neuron Formula

The output of a neuron is calculated as:

\[
\text{Output} = (x_1 \cdot w_1) + (x_2 \cdot w_2) + ... + (x_n \cdot w_n) + b
\]

Where:
- \(x_i\) = input
- \(w_i\) = corresponding weight
- \(b\) = bias

---

## Example: Neuron with 3 Inputs

```python
# Inputs: could be raw data or output from a previous layer
inputs = [1, 2, 3]

# Weights: trainable parameters
weights = [0.2, 0.8, -0.5]

# Bias: allows the neuron to shift the activation
bias = 2

# Output calculation
output = (
    inputs[0] * weights[0] +
    inputs[1] * weights[1] +
    inputs[2] * weights[2] +
    bias
)

print("Output (3 inputs):", output)


## A Layer of Neurons

````python
# Neural networks typically have layers that consists of more then one layer.
# Let's say we have a scenario with three neurons in a layer and 4 inputs.
````

````python
# Representation of the layer Below:
# https://www.youtube.com/watch?v=Uvngs6sWyBg
inputs = [1 , 2, 3, 2.5]

weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2
bias2 = 3
bias3 = 0.5

output = [
    # Neuron 1
    inputs[0]*weights1[0] +
    inputs[1]*weights1[1] +
    inputs[2]*weights1[2] +
    inputs[3]*weights1[3] + bias1,

    # Neuron 2
    inputs[0]*weights2[0] +
    inputs[1]*weights2[1] +
    inputs[2]*weights2[2] +
    inputs[3]*weights2[3] + bias2,

    # Neuron 3
    inputs[0]*weights3[0] +
    inputs[1]*weights3[1] +
    inputs[2]*weights3[2] +
    inputs[3]*weights3[3] + bias3,
    
] 

print(output)

"""
In this code, we have three sets of weights and three biases, which define three neurons. Each
neuron is “connected” to the same inputs. The difference is in the separate weights and bias
that each neuron applies to the input. This is called a fully connected neural network — every
in the current layer has connections to every neuron from the previous layer.
"""

#But this is not appropriate way to code or program 
````

````python
#we can iterate input with the help of loop and make things way easy.

# SAMPLE DATA
inputs = [1, 2, 3, 2.5]

weight = [[0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

# This will store the output of the Current layer.
layer_output = []

# This will iterate Weights and Biases into this variable(neuron_weights, neuron_bias).
for neuron_weights, neuron_bias in zip(weight, biases):
    
    # Zero output of the given neuron.
    neuron_output = 0
    
    # Here iteraion the WEIGHTS and INPUT. 
    for n_input, weight in zip(inputs, neuron_weights):

        # Multiplying input and weight
        neuron_output += n_input*weight

    # Now adding up the bias.
    neuron_output += neuron_bias

    # Adding up the output tp the Nueron Variable.
    layer_output.append(neuron_output)
print(layer_output)

````

# Matrix and Transpose of Matrix

##  What is a Matrix Product?
- A **matrix product** involves multiplying **two matrices**.
- It consists of computing the **dot product** between each **row of the first matrix** and **column of the second matrix**.
- This operation results in a **matrix of outputs**.

---

##  Matrix Product Requirements:
- If matrix **A** has shape `(m, n)` and matrix **B** has shape `(n, p)`, then the resulting matrix will have shape `(m, p)`.
- **Example:**
  - Matrix A: shape `(5, 4)`
  - Matrix B: shape `(4, 5)`
  - Result: shape `(5, 5)`

---

##  Dot Product vs Matrix Product:
- **Dot product** returns a single scalar value.
- **Matrix product** returns a matrix by applying dot product to rows and columns.

---

##  Vector-Based Matrix Product
- We can perform matrix products on **vectors** treated as matrices.

```python
a = [1, 2, 3]       # Row vector → shape (1, 3)
b = [[2], [3], [4]] # Column vector → shape (3, 1)

```

# Matrix Multiplication: Inputs × Weights (with Transposition)

## Key Concepts:

### 1. Why Transpose the Weights?
- The matrix product in a neural network involves:
- Inputs and weights are both stored as **row-aligned vectors**.
- To compute dot products correctly between each **input sample** and each **neuron’s weight set**, we must **transpose the weights matrix**.

---

### 2. When Transposition is Needed
- When `inputs` is a **single vector**, and `weights` is a matrix → dot product works directly.
- When `inputs` is a **batch** (matrix), and `weights` is also a matrix:
- We need to **transpose the weights** so the shapes align for matrix multiplication.

```python
# inputs: shape (batch_size, input_features)
# weights: shape (neurons, input_features)
# → transpose weights to shape (input_features, neurons)

````python
import numpy as np

inputs = [[1.0, 2.0, 3.0, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]

weights =[[0.2, 0.8, -0.5, 1.0],
          [0.5, -0.91, 0.17, 0.87],
          [-0.26, -0.27, 0.17, 0.87]]

biases = [2.0, 3.0, 0.5]

outputs = np.dot(inputs, np.array(weights).T) + biases

print(outputs)
````
